```python
import pandas as pd
df = pd.read_csv('學生成績.csv')
df
#dataFrame,依值使用subscript搜尋
#國文及格
df[df['國文'] >= 60]
```

```python
#數學不及格
df[df['數學'] < 60]
```

```python
#國文不及格同時英文不及格

(df['國文'] < 60) & (df['英文'] < 60)
```

```python
#國文不及格同時英文不及格

df[(df['國文'] < 60) & (df['英文'] < 60)]
```

```python
#全部都及格的
df[(df['國文'] >= 60) & (df['英文'] >= 60) & (df['數學'] >= 60) & (df['地理'] >= 60) & (df['歷史'] >= 60)]
```

```python
#平均80分以上的同學
df[df['平均'] >= 80]
```

```python
#國文不及格 或 英文不及格
df[(df['國文'] < 60) | (df['英文'] < 60)]
```
